Ejercicios de Limpieza

1. Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.

```python
data_final = data1.merge(data2, how="outer", left_on="Country", right_on="country")
data_final
```

```python
data_final.drop(columns= "country", inplace= True)
data_final
```

```python
data_final.columns
```

2. Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:
No tengan espacios.
Estén en minúscula.
No tengan paréntesis, es decir, quitar "(%)", "(Km2)".
Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.
Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

```python
columnas = data_final.columns
columnas

columnas_final = (
        columnas.str.lower()
        .str.replace(" ", "")
        .str.replace("\([^)]*\)", "", regex=True)
        .str.replace("\n", "")
        .str.replace(":", "")
)
columnas

data_final.columns = columnas_final
data_final
```

3. La columnas coordinates tiene la latitud y la longitud en una sola columna. 
Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates.

```python
data_final[['latitude', 'longitude']] = data_final['coordinates'].str.split(',', expand=True)
data_final.head(1)
```

4. Las columnas unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment, gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land tienen "%". Elimina los "%" de los valores de las columnas.

```python
lista = [
    "unemploymentrate",
    "totaltaxrate",
    "taxrevenue",
    "populationlaborforceparticipation",
    "outofpockethealthexpenditure",
    "grosstertiaryeducationenrollment",
    "grossprimaryeducationenrollment",
    "forestedarea",
    "cpichange",
    "agriculturalland",
]

for columnas_final in lista:
    data_final[columnas_final] = data_final[columnas_final].str.replace("%", "")
    
data_final.head(5).T
```

Ejercicios de Filtrado

1. Encuentra todos los países cuya mortalidad infantil esté entre 40 y 50 personas por kilómetro cuadrado.

```python
data_final.columns
```

```python
mor_infantil = data_final[(data_final['infantmortality'] >= 40) & (data_final['infantmortality'] <= 50)]
mor_infantil
```

2. Encuentra los países cuyas tasas de natalidad son mayores o iguales a 20 y su esperanza de vida es mayor de 75 años.

```python
tasa_nat = data_final[(data_final['birthrate'] >= 20) & (data_final['lifeexpectancy'] > 75)]
tasa_nat
```

3. Encuentra las ciudades cuyos paises contienen la palabra "la" en su nombre.

```python
ciudades_con_la = data_final[data_final['country'].str.lower().str.contains('la')]
ciudades_con_la
```

4. Encuentra los países cuyos medicos por cada 1000 habitantes (physicians_per_thousand) sea mayores de 5.

```python
paises_medicos_mayores_5 = data_final[data_final['physiciansperthousand'] > 5]
```

5. Encuentra los países cuyatasa de fertilidad sea mayor a 6.

```python
paises_fertilidad_mayor_6 = data_final[data_final['fertilityrate'] > 6]
paises_fertilidad_mayor_6
```

6. Encuentra los países cuya moneda es el euro (EUR) y tienen una tasa de natalidad superior al promedio.

```python
promedio_natalidad = data_final['birthrate'].mean()
paises_euro_y_natalidad_superior_promedio = data_final[(data_final['currency-code'] == 'EUR') & (data_final['birthrate'] > promedio_natalidad)]
paises_euro_y_natalidad_superior_promedio
```
